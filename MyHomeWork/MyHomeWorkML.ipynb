```python
import pandas as pd
prj_data = pd.read_csv("ai_training_t2_project_data.csv", sep=";")
email_data = pd.read_csv("email.csv")
```

```python
#Compreende-se que "Gabarito" do 1o CSV possui uma relação com "categoria" do 2o CSV. O que permite uma combinação. 
esquerda = prj_data.set_index(['Gabarito'])
direita = email_data.set_index(['categoria'])

esquerda.join(direita, lsuffix='_prj', rsuffix='_eml')
```
